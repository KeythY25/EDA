Task 2: Data Selection
The dataset that was choosen was Diamond Prices from Kaggle. The dataset comprises 53,940 diamonds and includes 10 distinct features. It also contains an unamed variable which considered to be the index for each diamond. The features are

- Carat: represents the weight or size of the diamond
- Cut: represents overall quality and value of a diamond. It has 5 distinct categories which is fair, good, very good, ideal, premium 
- Color: Determines a diamond's visual apperance and how it interact with light. D, E, F are colorless diamond while G, H, I, J is slightly tinted diamonds. The closer it is to color D the more colorless while closer to color J it is more tinted diamond
- Clarity: Determines the purity and transparency of a diamond
- Depth: Indicates the depth percentage of a diamond
- Table: Indicates the table facet to the average width of the diamond
- Price: Indicates the price of the diamond
- X: represents the length of the diamond in mm
- Y: represents the width of the diamond in mm 
- Z: represents the depth of the diamond in mm

While most of these features are numeric, it is important to note that the variables cut, color, and clarity are categorical factors with specific ordered levels



Task 3: Explortory Data Analysis

Conduct exploratory data analysis on the selected dataset using Python libraries (e.g.,  Pandas, Matplotlib, Seaborn). The analysis should include data cleaning, summary statistics, and visualizations to understand the data distribution and patterns.


```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import diamond dataset
diamonds = pd.read_csv('diamonds.csv')

# Display first few rows
print(diamonds.head())

# Data Cleaning
# Check for missing values
print("Missing values:")
print(diamonds.isnull().sum())

# Summary Statistics
print("Summary statistics:")
print(diamonds.describe())

# Data Visualization
# Pairplot to visualize relationships between numeric variables
sns.pairplot(data=diamonds, vars=['carat', 'depth', 'table', 'price'])
plt.show()

# Histogram of carat variables
plt.figure(figsize=(8, 4))
sns.histplot(diamonds['carat'], bins=30, color='skyblue')
plt.title('Distribution of Carat')
plt.xlabel('Carat')
plt.ylabel('Count')
plt.show()

# Boxplot of price by cut and color
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=diamonds, hue='color', palette='viridis')
plt.title('Price Distribution by Cut and Color')
plt.show()

# Count plot for cut variable
plt.figure(figsize=(8, 6))
sns.countplot(data=diamonds, x='cut', hue='cut', palette='Set1', legend=False)
plt.title('Distribution of Cut')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()


# Select only the numeric columns for correlation analysis
numeric_columns = diamonds.select_dtypes(include=['number'])

# Exclude the 'cut', 'color', and 'clarity' columns
columns_to_exclude = ['cut', 'color', 'clarity']
numeric_columns = numeric_columns.drop(columns_to_exclude, axis=1, errors='ignore')

# Create a correlation matrix for the selected columns
correlation_matrix = numeric_columns.corr()

# Create a correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix (Excluding "cut," "color," and "clarity")')
plt.show()

```

Insights: 
For the Exploratory Data Analysis, first we imported the libraries and dataset. Then printed out the first row of the data to see the variables that is going to be utilized in the research. Data Cleanup was the next step to ensure that there is no missing data in each variable. A statistical summary of of the dataset was then printed out. Then we utilized the maplotlib and Seaborn to create visualization that would show different correlation of the variables. First, a pairplot was created to show the correlation between the numerical variables, which carat, depth, table and price. In the second visual, it displays the distribution of carat, which has higher counts in between 0-1. The distribution can mean that there are more diamond that have size below 1 carat. The third visualization showed the price distribution with the regards to the diamonds cut and color. Fourth visualization is the count distribution of the five catergorical cuts, which shows that the Ideal cut has more count and fair has the lowest number. The last visulization shows the correlation between all numerical variables, which includes the length, depth, and width in millimeters.


Task 4: Research Questions, Hypothese, Correlation Analysis and Statistical Techniques

Research Question 1: Does the carat size of a diamond have a significant impact on its price?

Hypothesis 1: We hypothesize that there is a positive correlation between carat size and price, meaning that as the carat size increases, the price of the diamond also increases.

Research Question 2: Does the quality of the diamond's cut have a significant impact on its price?

Hypothesis 2: We hypothesize that diamonds with higher cut quality will have higher prices.

Research Question 3: How does the depth and table percentages of a diamond affect its price?

Hypothesis 3: We hypothesize that there may be a negative correlation between depth percentage and price, indicating that deeper diamonds are priced lower. Additionally, there may be a positive correlation between table percentage and price, suggesting that larger table percentages result in higher prices. 


```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import diamond dataset
diamonds = pd.read_csv('diamonds.csv')

# Check the first few rows of the dataset
print(diamonds.head())

# Data Cleaning
# Check for missing values
print("Missing values:")
print(diamonds.isnull().sum())

# Summary Statistics
print("Summary statistics:")
print(diamonds.describe())

# Research Question 1
# Correlation analysis for carat and price
correlation_carat_price = diamonds['carat'].corr(diamonds['price'])
print(f"Correlation between carat and price: {correlation_carat_price}")

# Scatter plot for carat vs. price
plt.figure(figsize=(8, 6))
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.5, color='b')
plt.title('Scatter Plot of Carat vs. Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()

plt.show()

# Research Question 2
# Calculate the mean price for each 'cut' category
mean_price_by_cut = diamonds.groupby('cut')['price'].mean()
print(mean_price_by_cut)

# Create a box plot to visualize the relationship between cut and price
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=diamonds, order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], palette='Set1', hue='cut', legend=False)
plt.title('Box Plot of Cut vs. Price')
plt.xlabel('Cut')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()

# Research Question 3
# Correlation analysis for depth, table, and price
correlation_depth_price = diamonds['depth'].corr(diamonds['price'])
correlation_table_price = diamonds['table'].corr(diamonds['price'])
print(f"Correlation between depth and price: {correlation_depth_price}")
print(f"Correlation between table and price: {correlation_table_price}")

# Pairplot for depth, table, and price
pairplot_data = diamonds[['depth', 'table', 'price']]
sns.pairplot(pairplot_data)
plt.show()

```

Insights:
Research Question 1: Carat size and price
    The correlation coefficient of 0.922 between carat size and price indicates a very strong positive linear relationship between the two variables. As carat size of a diamond increases, the price of the diamond tends to increases, menaing larger diamonds are generally more expensive.

Research Question 2: Cut vs Price
    The results shows the average prices of diamonds for the different cut. The average price of a "Premium" cut is the highest among all the cut categories, indicating that Premium cut diamonds tend to be more expensive on average. Ideal cut has the lowest average among the catergories suggesting that they are generally more affordable on average. The mean of each category displays that diamond's cut does have an impact on its price. Premium tends to command higher price while Fair cuts are typically less expensive on average.

Research Question 3: Depth, Table on Price
    The correlation between Depth and Price indicates a weak linear relationship. It is almost neutral suggesting that depth alone is not a strong predictor of price. The correlation of Table and price is slightly stronger than the correlation with depth, but it is still relatively weak.  Both results indicate that depth and table percentages are not strong determinants of diamond prices.

Task 5: Regression Model

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.compat import lzip

# Import diamond dataset
diamonds = pd.read_csv('diamonds.csv')

# Set carat and price for x and y variables
X = diamonds['carat']
y = diamonds['price']

# Add a constant to the independent variable
X = sm.add_constant(X)

# Fit a simple linear regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

# Assumption 1: Linearity
# Visualize the regression relationship
plt.figure(figsize=(10, 6))
sns.regplot(x='carat', y='price', data=diamonds, ci=None, scatter_kws={'alpha':0.3})
plt.title('Regression Relationship between Carat and Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()

# Assumption 2: Normality of Residuals
residuals = model.resid
plt.figure(figsize=(8, 6))
sns.histplot(residuals, kde=True, color='skyblue')
plt.title('Residuals Distribution')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()

# Assumption 3: Homoscedasticity
name = ['Lagrange multiplier statistic', 'p-value', 'f-value', 'f p-value']
test = het_breuschpagan(model.resid, model.model.exog)
lzip(name, test)
```

Interpretation:

    Assumptions:
    - Linearity: The regression plot suggests a linear relationship.
    - Normality of Residuals: The residuals appear to be approximately normally distributed.
    - Homoscedasticity: The Breusch-Pagan test results show whether the residuals have constant variance. A low p-value would indicate heteroscedasticity, while a high p-value suggests homoscedasticity.

    
    The regression modeling is between price and carat sizing. The R-squared value of 0.849 indicates that 84.9% of the variance in diamond prices is explained by the carat size. The F-statistics measures of the overall significance of the regression mode. In this model, the f-statistics is at 3.041e+05 which is very high and indicates it is statistically significant. Th coefficient table indicates the estimation relationship between price and carat sizing. The coefficient for carat indicates 7756.44 which menas that for eacch unit increase in carat the estimate increase in price is approximately 7756.44. The p-value of 0.000 indicates that the coeffiecients are highly significant. The Omnibus tests and Jarque-Bera suggests that the residuals are not normally distributed. The skew and kurtosis displayed that the model is positively skewed. The condition number suggest that there is multicollinearity in the model. Overall the high r-square and low p-values suggest that the model si a good for for the data, and carat variable is a significant predictor of diamond price. The Lagrange multiplier statistic indicates a high presence of heteroscedasticity. With the low p-value and high f-statistics indicates that the heteroscedasticity is statisticlaly significant, this implieas that the variance of residuals is not constant across all levels.
